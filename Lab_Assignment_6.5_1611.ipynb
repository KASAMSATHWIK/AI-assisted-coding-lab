```python
#Task 1: Conditional Eligibility Check (Voting System)


age = int(input("Enter your age: "))
citizen = input("Are you a citizen? (yes/no): ").lower()

if age >= 18 and citizen == "yes":
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")
```

```python
#Task 2: Loop-Based String Processing (Vowel & Consonant Count)

text = input("Enter a string: ").lower()

vowels = "aeiou"
vowel_count = 0
consonant_count = 0

for ch in text:
    if ch.isalpha():
        if ch in vowels:
            vowel_count += 1
        else:
            consonant_count += 1

print("Vowels:", vowel_count)
print("Consonants:", consonant_count)
```

```python
#Task 3: Library Management System (Classes + Loops + Conditions)

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, name):
        self.books.append(name)
        print("Book added.")

    def remove_book(self, name):
        if name in self.books:
            self.books.remove(name)
            print("Book removed.")
        else:
            print("Book not found.")

    def display_books(self):
        if not self.books:
            print("No books available.")
        else:
            print("Available Books:")
            for book in self.books:
                print("-", book)


library = Library()

while True:
    print("\n1. Add Book")
    print("2. Remove Book")
    print("3. Display Books")
    print("4. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        book = input("Enter book name: ")
        library.add_book(book)

    elif choice == 2:
        book = input("Enter book name: ")
        library.remove_book(book)

    elif choice == 3:
        library.display_books()

    elif choice == 4:
        print("Exiting...")
        break

    else:
        print("Invalid choice.")
```

```python
#Task 4: Class-Based Attendance System

class Attendance:
    def __init__(self):
        self.students = {}

    def mark_attendance(self, name, status):
        self.students[name] = status

    def display_attendance(self):
        print("\nAttendance Report:")
        for name, status in self.students.items():
            print(name, ":", status)


attendance = Attendance()

while True:
    print("\n1. Mark Attendance")
    print("2. Display Attendance")
    print("3. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        name = input("Enter student name: ")
        status = input("Present/Absent: ")
        attendance.mark_attendance(name, status)

    elif choice == 2:
        attendance.display_attendance()

    elif choice == 3:
        break

    else:
        print("Invalid choice.")
```

```python
#Task 5: ATM Menu (Loops + Conditionals

balance = 5000

while True:
    print("\n--- ATM MENU ---")
    print("1. Check Balance")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        print("Balance:", balance)

    elif choice == 2:
        amount = int(input("Enter deposit amount: "))
        balance += amount
        print("Amount deposited.")

    elif choice == 3:
        amount = int(input("Enter withdraw amount: "))

        if amount <= balance:
            balance -= amount
            print("Withdrawal successful.")
        else:
            print("Insufficient balance.")

    elif choice == 4:
        print("Thank you!")
        break

    else:
        print("Invalid option.")
```
